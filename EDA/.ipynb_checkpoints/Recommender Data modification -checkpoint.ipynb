# We need to merge different wide range of genre into broad genres for classification :

#### Below is the list of operations done to reduce it to the broad range of genres .

##### Fiction: Fiction, Crime, Mystery, Dark, Women's Fiction, Realistic fiction, Christian fiction, Young Adult, New Adult, Contemporary, Thriller, Suspense

#####   Non Fiction : Non Fiction, History, Historical, Media Tie in, Politics, Science, Travel, Biography, Autobiography,  Business.

#####   Children : Children . 

#####   Classics : Classics, Plays, Poetry

#####   Fantasy : Fantasy, Science fiction, Sequential art,  Paranormal

#####   Romance : Romance, Erotica, Feminism

#####   Miscellaneous : 40k, Academic, Criticism, Economics, Education, Food and drinks, Humor, Leadership, Lgbt,  Philosophy, Parenting,  

#####       > .                         Psychology,  Self Help, Short Stories, Writing



k =['Fiction', 'Nonfiction', 'Childrens', 'Biography', 'History',
       'Fantasy', 'Sequential Art', 'Autobiography', 'Historical',
       'Short Stories', 'Media Tie In', 'Travel', 'Self Help', 'Business',
       'Psychology', 'Romance', 'Science', 'Politics', 'Leadership',
       'Economics', 'Realistic Fiction', 'Poetry', 'Animals', 'Mystery',
       'Classics', 'Science Fiction', 'Plays', 'Young Adult', 'Academic',
       'Horror', 'Philosophy', 'Paranormal', 'Womens Fiction',
       'Contemporary', 'Thriller', 'Christian Fiction', 'Crime',
       'New Adult', 'Dark', 'Erotica', 'Suspense', 'Anthologies', '40k',
       'Lgbt', 'Parenting', 'Humor', 'genre', 'Writing', 'Feminism',
       'Education', 'Criticism', 'Food and Drink']

```python
f['genre']=f['genre'].replace(to_replace ={'40k',"Academic",'Criticism','Economics','Education','Food and Drink',
                                          'Humor','Leadership','Lgbt','Philosophy','Parenting','Psychology',
                                           'Self Help','Short Stories','Writing'},value ='Miscellaneous',regex =True)
```

```python
f['genre']=f['genre'].replace(to_replace = {'Crime','Mystery','Womens Fiction','Christian Fiction','Realistic Fiction'
                                           ,'Young Adult','New Adult','Contemporary','Thriller','Suspense','Dark'
                                           ,'Anthologies',' Non fiction Fiction'},
                                      value='Fiction', regex = True
                             )
```

```python
f['genre']=f['genre'].replace(to_replace={'Plays','Poetry'}, value ="Classics",regex =True)
```

```python
f['genre']=f['genre'].replace(to_replace={'Nonfiction','History','Historical','Media Tie In','Politics'
                                         ,'Science','Travel','Biography','Autobiography','Non fiction Fiction'
                                         'Business','Academic','Business'}, value =" Non fiction",regex =True)
```

```python
f['genre']=f['genre'].replace(to_replace ={'Erotica','Feminism','Romance'},value ='Romance',regex =True)
```

```python
f['genre']=f['genre'].replace(to_replace={'Fantasy','Science fiction',
                                          'Sequential Art','Paranormal'},value ='Fantasy',regex =True )
```

```python
f['genre']=f['genre'].replace(to_replace = {'Animals','Childrens'},value ='Children',regex =True )
```

### Dropping the genres which have unicode conversion errors etc 

```python
f=f[f.genre != "Horror"]
f=f[f.genre != 'genre']
# f=f[f.genre != ' Non fiction Fiction']
```
